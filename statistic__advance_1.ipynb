Question 2: What are the types of random variables?

Answer:
Types of Random Variables
A random variable is a numerical value that represents the outcome of a random process.
It can be classified based on the nature of the values it takes.

1. Discrete Random Variable
Definition:
A random variable that takes a finite or countably infinite set of values.

Characteristics:

Values are separate and distinct (no fractions between them).

Probabilities are described using a Probability Mass Function (PMF).

The sum of probabilities of all possible values equals 1.

Examples:

Number of heads in 3 coin tosses → {0, 1, 2, 3}.

Number of customers in a shop in an hour.

2. Continuous Random Variable
Definition:
A random variable that can take any value within a given range (infinite possibilities).

Characteristics:

Values are not countable.

Probabilities are described using a Probability Density Function (PDF).

The probability of a specific single value is zero; instead, probabilities are calculated over an interval.

Examples:

Height of students (can be 165.5 cm, 165.55 cm, etc.).

Time taken to complete a task.

3. Mixed Random Variable (Less Common)
Definition:
A variable that has both discrete and continuous components.

Examples:

Time until a bus arrives (continuous), but the number of buses arriving in an hour (discrete).

(Sample Space → Random Variable → Probability Distribution) can be drawn to illustrate both types.

Importance of Knowing the Types
Helps in choosing the right statistical methods.

Essential for probability modeling, data analysis, and machine learning.



Question 3: Explain the difference between discrete and continuous distributions.

1. Definition

Discrete Distribution: Represents data that can take only specific, countable values.

Continuous Distribution: Represents data that can take any value within a given range (including fractions/decimals).

Discrete: Values are separate and distinct (no in-betweens).

Question 9: Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. Compute its mean and standard
deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from N(50, 5)
data = np.random.normal(loc=50, scale=5, size=1000)

# Step 2: Compute mean and standard deviation
mean_value = np.mean(data)
std_dev_value = np.std(data)

print(f"Sample Mean: {mean_value:.2f}")
print(f"Sample Standard Deviation: {std_dev_value:.2f}")

# Step 3: Draw histogram
plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Normally Distributed Data (Mean=50, SD=5)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Draw a vertical line for mean
plt.axvline(mean_value, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean_value:.2f}")
plt.legend()

plt.show()
```

Question 10: You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.

```python
'''Part 1 — Concept & Explanation
The Central Limit Theorem (CLT) states that if we take repeated random samples from a population and calculate their means, the sampling distribution of the mean will approximate a normal distribution, regardless of the population’s original shape, as long as the sample size is large enough (usually n ≥ 30).

For estimating average daily sales:

Why CLT matters here

Even though we only have 20 daily sales values in this dataset, in real scenarios we could collect multiple samples over time.

The CLT allows us to use normal distribution-based methods to estimate a confidence interval for the mean sales.

Steps to Apply CLT

Step 1: Calculate the sample mean (x̄) and sample standard deviation (s).

Step 2: Compute the standard error (SE):

𝑆
𝐸
=
𝑠
𝑛
SE=
n
​

s
​

Step 3: For a 95% Confidence Interval (CI), use the t-distribution (because n < 30).
Formula:

𝐶
𝐼
=
𝑥
ˉ
±
𝑡
(
𝑛
−
1
,
𝛼
/
2
)
×
𝑆
𝐸
CI=
x
ˉ
 ±t
(n−1,α/2)
​
 ×SE
where
𝑡
(
𝑛
−
1
,
0.025
)
t
(n−1,0.025)
​
  is the critical t-value from the t-table.

Interpretation

The CI tells us a range where the true average daily sales is likely to fall 95% of the time.'''


import numpy as np
import scipy.stats as st

# Daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to numpy array
data = np.array(daily_sales)

# Step 1: Mean and Standard Deviation
mean_sales = np.mean(data)
std_sales = np.std(data, ddof=1)  # sample standard deviation

# Step 2: Sample size
n = len(data)

# Step 3: Standard Error
SE = std_sales / np.sqrt(n)

# Step 4: t-critical value for 95% confidence
t_critical = st.t.ppf(0.975, df=n-1)

# Step 5: Margin of Error
margin_of_error = t_critical * SE

# Step 6: Confidence Interval
lower_bound = mean_sales - margin_of_error
upper_bound = mean_sales + margin_of_error

# Output
print(f"Mean Daily Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")



```
